```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# scrapeo de tabla de alex volkanovski
URL='https://en.wikipedia.org/wiki/Alexander_Volkanovski'

html = req.get(URL).text
sopa = bs(html, 'html.parser')
```

```python
tabla2 = sopa.find_all('table')[2]
tabla2
```

```python
cabeceras2=[]

for e in tabla2.find_all('th'):
    
    cabeceras2.append(e.text.strip())

cabeceras2
```

```python
filas2 = tabla2.find_all('tr')
```

```python
data2 = []

for f in filas2:
    
    tmp2= []
    
    for e in f.find_all('td'):
        
        tmp2.append(e.text.strip())
        
    data2.append(tmp2)
        
data2
```

```python
#ahora pasados los datos de Alexander VOlkanowvski a dataframe. Vamos a limpiar.

alex_history=pd.DataFrame(data2, columns=cabeceras2)

alex_history
```

```python
alex_history['Method']=alex_history['Method'].str.replace(r'\(.*\)', '', regex=True).str.strip()
alex_history
```

```python
alex_history['Method'] = alex_history['Method'].str.split(':', expand=True)[0].str.strip()
alex_history
```

```python
column_name_to_remove = 'Notes'
alex_history = alex_history.drop(column_name_to_remove, axis=1)
```

```python
alex_trilogy = alex_history[alex_history['Opponent'].str.contains('Max Holloway', na=False)]
alex_trilogy
```

```python
alex_trilogy['Location_id']=[1,2,3]
```

```python
alex_trilogy.rename(columns={'Res.': 'Res'}, inplace=True)
```
